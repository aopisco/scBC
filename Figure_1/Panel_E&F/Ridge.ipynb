```python
import scanpy as sc
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
adata_processed = sc.read_h5ad('SS2_processed.h5ad')
```

```python
adata_processed
```

```python
df_loadings = pd.DataFrame(adata_processed.obsm['X_pca'], index=adata_processed.obs_names)
```

```python
for i in df_loadings.columns:
    df_loadings = df_loadings.rename(columns={i: "pc_"+str(i)})
df_loadings
```

```python
auxdict = pd.Series(adata_processed.obs['ER_receptor'].values,index=adata_processed.obs.index).to_dict()
df_loadings['ER_receptor'] = df_loadings.index.map(auxdict)
auxdict = pd.Series(adata_processed.obs['intrinsic_BC_type'].values,index=adata_processed.obs.index).to_dict()
df_loadings['intrinsic_BC_type'] = df_loadings.index.map(auxdict)
auxdict = pd.Series(adata_processed.obs['sort'].values,index=adata_processed.obs.index).to_dict()
df_loadings['sort'] = df_loadings.index.map(auxdict)
df_loadings
```

```python
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
g = sns.FacetGrid(df_loadings, row="ER_receptor", height=1.2,hue="ER_receptor",palette=['#02818a','#810f7c'], hue_kws={"alpha": [0.9,0.9]},size=1,aspect=3)
g.map_dataframe(sns.kdeplot, x="pc_0", fill=True, alpha=1)
g.map_dataframe(sns.kdeplot, x="pc_0", color='black')
g.fig.subplots_adjust(hspace=-.5)
g.set_titles("")
g.set(yticks=[],xticks=[])
g.despine(left=True)
plt.savefig('Ridge_SS2_ER.pdf',bbox_inches='tight')
```

```python
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
g = sns.FacetGrid(df_loadings, row="intrinsic_BC_type", height=1.2,hue="intrinsic_BC_type",palette=['#9ebcda','#8856a7'], hue_kws={"alpha": [0.9,0.9]},size=1,aspect=3)
g.map_dataframe(sns.kdeplot, x="pc_0", fill=True, alpha=1)
g.map_dataframe(sns.kdeplot, x="pc_0", color='black')
g.fig.subplots_adjust(hspace=-.5)
g.set_titles("")
g.set(yticks=[],xticks=[])
g.despine(left=True)
plt.savefig('Ridge_SS2_subtype.pdf',bbox_inches='tight')
```
