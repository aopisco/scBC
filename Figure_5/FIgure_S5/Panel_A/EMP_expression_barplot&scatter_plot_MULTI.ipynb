```python
import scanpy as sc
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
adata_MULTI_processed = sc.read_h5ad('MULTI_processed.h5ad')
```

```python
adata_MULTI_processed.X = adata_MULTI_processed.layers['log1p']
```

```python
MULTI_log_counts_df = adata_MULTI_processed.to_df()
```

```python
for i in MULTI_log_counts_df.index:
    MULTI_log_counts_df.loc[i, 'EMP_score'] = adata_MULTI_processed.obs.loc[i, 'EMP_score']
    MULTI_log_counts_df.loc[i, 'EMP_stage'] = adata_MULTI_processed.obs.loc[i, 'EMP_stage']
```

```python
MULTI_log_counts_df
```

```python
overlap_H_genes = ['CRYAB', 'KRT15', 'S100A2', 'CD24', "CALML5"]
```

# making scatter plots

```python
# for MULTI-seq

for i in overlap_H_genes:
    if i in MULTI_log_counts_df.columns:
        MULTI_log_counts_df_filtered = MULTI_log_counts_df[MULTI_log_counts_df[i] > 0]
        f, ax = plt.subplots(figsize=(2,2)) 
        sns.regplot(x='EMP_score', y=i, data=MULTI_log_counts_df_filtered,ax=ax,order=2,scatter=False,color='black', ci=None)
        sns.scatterplot(x='EMP_score', y=i, data=MULTI_log_counts_df, alpha=0.5, legend='full',ax=ax,hue='EMP_stage',hue_order = ['Epithelial-like', 'EMP Intermediate', 'Mesenchymal-like'],
               palette=['#81B1D3','#BC80B7','#F47F72'],s=5,edgecolor=None)
        ax.set_title('gene: '+i)
        ax.set_xlim(-1.2,1.2)
        ax.set_ylim(-0.1,)
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.savefig('MULTI_scatter_'+i+'_ref.pdf', bbox_inches="tight")
        plt.show()
        
    
        f, ax = plt.subplots(figsize=(2,2)) 
        sns.regplot(x='EMP_score', y=i, data=MULTI_log_counts_df_filtered,ax=ax,order=2,scatter=False,color='black', ci=None)
        sns.scatterplot(x='EMP_score', y=i, data=MULTI_log_counts_df, alpha=0.5, legend='full',ax=ax,hue='EMP_stage',hue_order = ['Epithelial-like', 'EMP Intermediate', 'Mesenchymal-like'],
               palette=['#81B1D3','#BC80B7','#F47F72'],s=5,edgecolor=None)
        ax.set_xlim(-1.2,1.2)
        ax.set_ylim(-0.1,)
        plt.legend([],[],frameon=False)
        ax.set(xlabel=None,ylabel=None,xticklabels=[],yticklabels=[])
        plt.savefig('MULTI_scatter_'+i+'.pdf', bbox_inches="tight")
        plt.show()
        print("-"*100)
```

# making bar plots

```python
# for MULTI-seq

for i in overlap_H_genes:
    if i in MULTI_log_counts_df.columns:
        temp_df = pd.DataFrame(index=set(MULTI_log_counts_df['EMP_stage']),columns=['fraction'])
        for j in set(MULTI_log_counts_df['EMP_stage']):
            temp_EMT = MULTI_log_counts_df[MULTI_log_counts_df['EMP_stage']==j]
            temp_EMT_filtered = temp_EMT[temp_EMT[i] > 0]
            temp_df.loc[j, 'fraction'] = len(temp_EMT_filtered)/len(temp_EMT)
            temp_df['EMP_stage'] = temp_df.index
        f, ax = plt.subplots(figsize=(2,0.5))
        sns.barplot(x="EMP_stage", y="fraction", data=temp_df,order = ['Epithelial-like', 'EMP Intermediate', 'Mesenchymal-like'],
               palette=['#81B1D3','#BC80B7','#F47F72'])
        ax.set_ylim(0,1)
        plt.yticks([0, 0.5, 1])
        ax.set_title('gene: '+i)
        ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
        plt.savefig('MULTI_bar_plots_'+i+'_ref.pdf', bbox_inches="tight")
        plt.show()
        
        
        f, ax = plt.subplots(figsize=(2,0.5))
        sns.barplot(x="EMP_stage", y="fraction", data=temp_df,order = ['Epithelial-like', 'EMP Intermediate', 'Mesenchymal-like'],
               palette=['#81B1D3','#BC80B7','#F47F72'])
        ax.set_ylim(0,1)
        plt.yticks([0, 0.5, 1])
        ax.axes.xaxis.set_visible(False)
        plt.legend([],[],frameon=False)
        ax.set(xlabel=None,ylabel=None,xticklabels=[],yticklabels=[])
        plt.savefig('MULTI_bar_plots_'+i+'.pdf', bbox_inches="tight")
        plt.show()
        print("-"*100)
```
