## making h5ad by loading metadata

```python
# Importing the metadata

meta_data = pd.read_csv('files/scBC_ss2_metadata.csv',index_col='Plate_ID')
```

```python
meta_data
```

```python
# Creating Scanpy anndata object with combining gene cell table and metadata

adata = sc.AnnData(gc_table)

adata.obs['plate_ID'] = [i.split('_')[1] for i in adata.obs.index] 
adata.obs['well'] = [i.split('_')[0] for i in adata.obs.index]
adata.obs['cell_id'] = [(i.split('_')[0] + "_" + i.split('_')[1]) for i in adata.obs.index]
myList = list(set(adata.obs['plate_ID']))

try:
    auxdict = dict(meta_data['PDX_line'][myList])
    adata.obs['Tumor_ID'] = adata.obs['plate_ID'].map(auxdict)
    
    auxdict = dict(meta_data['Animal_ID'][myList])
    adata.obs['Animal_ID'] = adata.obs['plate_ID'].map(auxdict)
    
    auxdict = dict(meta_data['Receptor status'][myList])
    adata.obs['receptor_status'] = adata.obs['plate_ID'].map(auxdict)
    
    auxdict = dict(meta_data['ER receptor'][myList])
    adata.obs['ER_receptor'] = adata.obs['plate_ID'].map(auxdict)
    
    auxdict = dict(meta_data['intrinsic BC type'][myList])
    adata.obs['intrinsic_BC_type'] = adata.obs['plate_ID'].map(auxdict)
    
    auxdict = dict(meta_data['metastatic potential rank'][myList])
    adata.obs['metastatic_potential_rank'] = adata.obs['plate_ID'].map(auxdict)

    auxdict = dict(meta_data['metastatic potential group'][myList])
    adata.obs['metastatic_potential_group'] = adata.obs['plate_ID'].map(auxdict)
    
    for i in list(adata.obs.index):
        if ord('A') <= ord(adata.obs.loc[i, 'well'][0]) <= ord('H'):
            adata.obs.loc[adata.obs.index==i, 'sort'] = 'Tumor'
        elif ord('I') <= ord(adata.obs.loc[i, 'well'][0]) <= ord('P'):
            adata.obs.loc[adata.obs.index==i, 'sort'] = 'Metastatic'
    
    print('import complete for run ')
except KeyError:
    print('sorry, no metadata available for run')
```

```python
sc.pp.filter_cells(adata, min_counts=0)
sc.pp.filter_cells(adata, min_genes=0)
sc.pp.filter_genes(adata, min_cells=0)
```

```python
# calculating percent ERCC

ERCC_genes = adata.var_names.str.startswith('ERCC-')
adata.obs['percent_ERCC'] = np.sum(
    adata[:, ERCC_genes].X, axis=1) / np.sum(adata.X, axis=1) * 100
```

```python
# calculating percent MITO

mito_genes = adata.var_names.str.startswith('MT-')
adata.obs['percent_mito'] = np.sum(
    adata[:, mito_genes].X, axis=1) / np.sum(adata.X, axis=1) * 100
```

```python
# calculating percent Ribo

RP_genes = adata.var_names.str.startswith(("RPS","RPL"))
adata.obs['percent_ribo'] = np.sum(
    adata[:, RP_genes].X, axis=1) / np.sum(adata.X, axis=1) * 100
```
